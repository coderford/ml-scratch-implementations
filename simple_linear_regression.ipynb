## Problem

Write simple linear regression algorithm (1 independent var, 1 dependent var)

- **Input**: x, y
- **Output**: Regressor/Predictor

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Data setup
num_datapoints = 15
x = np.random.uniform(high=10, size=num_datapoints)
y = 5 * x + 3 + np.random.normal(loc=0, scale=3, size=num_datapoints)
```

```python
# Plot data
plt.scatter(x, y)
plt.show()
```

## Solution/Implementation

```python
x_bar = np.mean(x)
y_bar = np.mean(y)
m = np.sum(np.dot((x - x_bar), (y - y_bar)))  /  np.sum(np.square(x - x_bar))
c = y_bar - m * x_bar
```

```python
y_pred = m * x + c
```

```python
# plot results
plt.scatter(x, y)
plt.plot(x, y_pred, color='black')
plt.show()
```
